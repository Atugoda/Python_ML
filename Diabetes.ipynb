```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load the dataset
data = pd.read_csv(r'C:\Users\LENOVO\Downloads\Diabetes.csv')

# Handle missing values
missing_count = data.isna().sum().sum()
data = data.fillna(data.mean())
print(f"Handled {missing_count} missing values using the mean.")

# Split the data into features and target
X = data.drop('Class', axis=1)
y = data['Class']

# Train Logistic Regression model
lr_model = LogisticRegression()
lr_model.fit(X, y)
lr_pred = lr_model.predict(X)
lr_accuracy = accuracy_score(y, lr_pred)
lr_report = classification_report(y, lr_pred)
lr_confusion = confusion_matrix(y, lr_pred)
print('Logistic Regression:')
print(f'Accuracy: {lr_accuracy:.2f}')
print(classification_report(y, lr_pred))
print(confusion_matrix(y, lr_pred))

# Train SVM models with different kernels
for kernel in ['linear', 'rbf', 'poly']:
    svm_model = SVC(kernel=kernel)
    svm_model.fit(X, y)
    svm_pred = svm_model.predict(X)
    svm_accuracy = accuracy_score(y, svm_pred)
    svm_report = classification_report(y, svm_pred)
    svm_confusion = confusion_matrix(y, svm_pred)
    print(f'SVM with {kernel} kernel:')
    print(f'Accuracy: {svm_accuracy:.2f}')
    print(classification_report(y, svm_pred))
    print(confusion_matrix(y, svm_pred))

# Train Random Forest model
rf_model = RandomForestClassifier()
rf_model.fit(X, y)
rf_pred = rf_model.predict(X)
rf_accuracy = accuracy_score(y, rf_pred)
rf_report = classification_report(y, rf_pred)
rf_confusion = confusion_matrix(y, rf_pred)
print('Random Forest:')
print(f'Accuracy: {rf_accuracy:.2f}')
print(classification_report(y, rf_pred))
print(confusion_matrix(y, rf_pred))
```

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Load the digits dataset
digits = load_digits()
print(f"images shape: {digits.images.shape}")
print(f"Labels shape: {digits.target.shape}")

# Reshape the data
X = digits.images.reshape((len(digits.images), -1))
y = digits.target

# Standardize the data
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Logistic Regression model
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
lr_accuracy = accuracy_score(y_test, lr_pred)

# Train SVM model
svm_model = SVC(kernel='rbf', gamma=0.001, C=1.0)
svm_model.fit(X_train, y_train)
svm_pred = svm_model.predict(X_test)
svm_accuracy = accuracy_score(y_test, svm_pred)

# Train Random Forest model
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_pred)

# Evaluate the models
print("Logistic Regression:")
print(f"Accuracy: {lr_accuracy:.2f}")
print(classification_report(y_test, lr_pred))
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, lr_pred), annot=True, cmap='Blues', fmt='g')
plt.title('Logistic Regression Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

print("\nSVM:")
print(f"Accuracy: {svm_accuracy:.2f}")
print(classification_report(y_test, svm_pred))
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, svm_pred), annot=True, cmap='Blues', fmt='g')
plt.title('SVM Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

print("\nRandom Forest:")
print(f"Accuracy: {rf_accuracy:.2f}")
print(classification_report(y_test, rf_pred))

# Calculate Gini index for Random Forest
feature_importances = rf_model.feature_importances_
gini_index = 1 - (feature_importances ** 2).sum()
print(f"Gini index: {gini_index:.2f}")

plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, rf_pred), annot=True, cmap='Blues', fmt='g')
plt.title('Random Forest Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load the digits dataset
digits = load_digits()
print(f"Images shape: {digits.images.shape}")
print(f"Labels shape: {digits.target.shape}")

# Introduce missing values artificially for demonstration
X = digits.images.reshape((len(digits.images), -1))
y = digits.target
np.random.seed(42)
missing_mask = np.random.rand(*X.shape) < 0.1  # 10% missing values
X[missing_mask] = np.nan

# Display missing values count
print("\nMissing Values Per Feature (Before Handling):")
print(np.isnan(X).sum(axis=0))

# Handle missing values by replacing with column means
col_means = np.nanmean(X, axis=0)
inds = np.where(np.isnan(X))
X[inds] = np.take(col_means, inds[1])

print("\nMissing Values Per Feature (After Handling):")
print(np.isnan(X).sum(axis=0))

# Standardize the data
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Logistic Regression
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
lr_accuracy = accuracy_score(y_test, lr_pred)
print("\nLogistic Regression:")
print(f"Training Accuracy: {lr_model.score(X_train, y_train):.2f}")
print(f"Test Accuracy: {lr_accuracy:.2f}")
print("Confusion Matrix:")
print(confusion_matrix(y_test, lr_pred))

# SVM with RBF Kernel
svm_model = SVC(kernel='rbf', gamma=0.001, C=1.0)
svm_model.fit(X_train, y_train)
svm_pred = svm_model.predict(X_test)
svm_accuracy = accuracy_score(y_test, svm_pred)
print("\nSVM (RBF Kernel):")
print(f"Training Accuracy: {svm_model.score(X_train, y_train):.2f}")
print(f"Test Accuracy: {svm_accuracy:.2f}")
print("Confusion Matrix:")
print(confusion_matrix(y_test, svm_pred))

# Random Forest Classifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_pred)
print("\nRandom Forest:")
print(f"Training Accuracy: {rf_model.score(X_train, y_train):.2f}")
print(f"Test Accuracy: {rf_accuracy:.2f}")
print("Confusion Matrix:")
print(confusion_matrix(y_test, rf_pred))

# Evaluate feature importances and Gini index for Random Forest
feature_importances = rf_model.feature_importances_
gini_index = 1 - (feature_importances ** 2).sum()
print(f"\nRandom Forest Gini Index: {gini_index:.2f}")
```
